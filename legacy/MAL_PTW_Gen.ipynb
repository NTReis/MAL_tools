```python
import xml.etree.ElementTree as ET
import random

# === CONFIGURATION ===
FILE_PATH = "/content/animelist_1741818669_-_4850763.xml"  # Path to your MAL export

# === PARSE XML AND BUILD DICTIONARY ===
def parse_mal_xml(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()

    anime_dict = {}
    for anime in root.findall("anime"):
        status = anime.find("my_status").text.strip()
        title = anime.find("series_title").text.strip()
        anime_type = anime.find("series_type").text.strip()  # TV, OVA, Movie, etc.
        episodes = int(anime.find("series_episodes").text.strip())  # Convert to int
        id = anime.find("series_animedb_id").text.strip()

        if status == "Plan to Watch":  # Only include "Plan to Watch" anime
            anime_dict[title] = {
                "type": anime_type,
                "episodes": episodes,
                "id": id
            }

    return anime_dict
```

```python


# === FILTER FUNCTION ===
def filter_anime(anime_dict, min_episodes=None, max_episodes=None, anime_type=None):
    filtered = {}

    for title, details in anime_dict.items():
        if anime_type and details["type"] != anime_type:
            continue  # Skip if it doesn't match the type filter

        if min_episodes and details["episodes"] < min_episodes:
            continue  # Skip if below min episodes

        if max_episodes and details["episodes"] > max_episodes:
            continue  # Skip if above max episodes

        filtered[title] = details

    return filtered

# === PICK RANDOM ANIME ===
def pick_random_anime(anime_dict):
    if not anime_dict:
        return "No anime found with the given filters."

    title, details = random.choice(list(anime_dict.items()))
    return f"🎲 Random Pick: {title} ({details['type']}, {details['episodes']} episodes) https://myanimelist.net/anime/{details['id']}/"

# === RUN THE SCRIPT ===
anime_dict = parse_mal_xml(FILE_PATH)

# === APPLY FILTERS ===
filtered_anime = filter_anime(anime_dict, min_episodes=5, max_episodes=12, anime_type="TV")

# === PICK A RANDOM ANIME ===
print(pick_random_anime(filtered_anime))
```
